```python
import censusgeocode
import pandas as pd
import numpy as np
```

```python
crimes_df = pd.read_excel('2020crimes.xlsx')
```

```python
crimes_df['Street\nType'] = crimes_df['Street\nType'].replace(np.nan, '', regex=True)

crimes_df = crimes_df.astype(str)

crimes_df
```

```python
crimes_df['address_number'] = crimes_df['Block Range'].apply(lambda x: x.split('-')[0])
```

```python
crimes_df['address'] = crimes_df['address_number'] + ' ' + crimes_df['StreetName'] + ' ' + crimes_df['Street\nType']

crimes_df
```

```python
crimes_df['state'] = 'TX'

crimes_df
```

```python
crimes_df.rename(columns = {'Incident':'unique id', 'address':'street address', 'City':'city', 'ZIP Code':'zip code'}, inplace=True)

crimes_df.head()
```

```python
geo_df = crimes_df[['street address', 'city', 'zip code', 'state']]

geo_df
```

```python
import numpy as np
geo_df = geo_df.replace(np.nan, '', regex=True)
geo_df
```

```python
geo_df[:10].to_csv('2020crimes_first.csv', header=False)
```
